```python
from keras.layers import Input,Dense,Flatten
from keras.models import Model
#getting the pretrained model
from keras.applications.vgg16 import VGG16,preprocess_input # Vgg expects BGR images - so we have the reshape RGB channels


from keras.preprocessing import image 
from keras.preprocessing.image import ImageDataGenerator
from keras import utils

from sklearn.metrics import confusion_matrix

import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
image_size=[100,100]
```

```python
epochs=5
batch_size=32
```

```python
train_path='/content/data/fruits-360-original-size/fruits-360-original-size/Training'
valid_path='/content/data/fruits-360-original-size/fruits-360-original-size/Validation'
```

```python
image_files=glob(train_path+'/*/*.jp*g')
valid_files=glob(valid_path+'/*/*.jp*g')
```

```python
#build model 
```

```python
vgg=VGG16(input_shape=image_size+[3],weights='imagenet',include_top=False) # when include top is False then we are not using feature classifier
```

```python
vgg.summary()
```

```python
vgg=VGG16(input_shape=(224,224,3),weights='imagenet',include_top=True)
```

```python
vgg.summary()
```

```python
for layer in vgg.layers:
  layer.trainable=False
```

```python
vgg.summary()
```

```python
x=Flatten()(vgg.output)
x=Dense(len(folders),activation='softmax')(x)
```

```python
model1=Model(vgg.input,x)
```

```python
model1.summary()
```

```python
model1.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics='accuracy')
```

```python
gen=ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    shear_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    preprocessing_function=preprocess_input

)
```

```python
test_gen=gen.flow_from_directory(valid_path,target_size=image_size)
print(test_gen.class_indices)
```

```python
labels=[None]*len(test_gen.class_indices)
for k,v in test_gen.class_indices.items():
  labels[v]=k
```

```python
train_generator=gen.flow_from_directory(train_path,target_size=image_size,shuffle=True
                                        ,batch_size=batch_size)
valid_generator=gen.flow_from_directory(valid_path,target_size=image_size,shuffle=True
                                        ,batch_size=batch_size)
```

```python
r=model1.fit(train_generator,validation_data=valid_generator, epochs=1,steps_per_epoch=len(image_files)//batch_size, validation_steps=len(valid_files)//batch_size)
```
